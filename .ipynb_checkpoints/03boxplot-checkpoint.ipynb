```python
import seaborn as sns
import matplotlib.pyplot as plt
# import pandas as pd
# import numpy as np
# sns.theme()
kashti = sns.load_dataset("titanic")
kashti.head()
```

```python

p1= sns.boxplot(x= "sex" , y= "class", data=kashti,order=["female","male"],hue="alive")
# plt.show()
```

```python
sns.boxplot(x= "sex" , y= "class", data=kashti,order=["female","male"],hue="alive",showmeans= True)
```

```python
## This line of code is the purpose of mean and median first of all you are writing showmeans True
sns.boxplot(x= "sex" , y= "class", data=kashti,order=["female","male"],hue="alive",showmeans= True,
            meanprops={"marker":"*","markersize":"10","markeredgecolor":"red"})
## show labels 
plt.xlabel("male is survived"),
plt.ylabel("Age (in month)"),
plt.title("This is the my Box Plot")
plt.show()
```
